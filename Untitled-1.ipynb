```python
X = df.drop(columns=['Outcome'])  # Features
y = df['Outcome']  
#X = std.fit_transform(X)
```

```python
X_trian, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
model = XGBClassifier(
    objective='binary:logistic',  # Binary classification task
    eval_metric='logloss',        # Logarithmic loss as the evaluation metric
    use_label_encoder=False,      # Disable label encoding warning
    max_depth=3,                  # Depth of the decision tree
    learning_rate=0.1,            # Learning rate
    n_estimators=100,             # Number of boosting rounds
    random_state=0,               # Random seed for reproducibility
    verbosity=2
)
print(model)
model.fit(X_trian,y_train,verbose=1)
```

```python
model.fit(X_trian,y_train)

#warnings.warn(smsg, UserWarning)
```

```python
y_pred = model.predict(X_test)
#y_pred_proba = model.predict_proba(X_test)[:, 1]
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision=precision_score(y_test,y_pred)
recall=recall_score(y_test,y_pred)
f1=f1_score(y_test,y_pred)
roc_auc=roc_auc_score(y_test,y_pred)
conf_metrix=confusion_matrix(y_test,y_pred)
```

```python
print("Accuracy:", accuracy)
print("precision:",precision)
print("recall:",recall)
print("F1_score:", f1)
print("roc:",roc_auc)
print("confusion_matrix:",conf_metrix)
```

```python
sns.heatmap(conf_metrix, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.barh(X.columns, model.feature_importances_)
plt.title('Feature Importance')
plt.xlabel('Importance Score')
plt.ylabel('Features')
plt.show()
```
